```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', None)
```

# Miniscreen05

```python
benchmark = pd.read_excel("MiniScreen05_Count_Results.xlsx", index_col=0).iloc[:80, :]
```

```python
def ratio_benchmark_heatmap(file):
    df = pd.read_csv(file, index_col=0)
    filter_df = df[np.mod(np.arange(1, df.index.size + 1),9)!=0].iloc[:80, 1:]
    filter_df.index = benchmark.index
    filter_df.rename(columns={'Vec':'Vector'}, inplace=True)
    
    test = filter_df.div(benchmark)

    sns.set(font_scale=0.8)
    fig, ax = plt.subplots(figsize=(20,20))
    heatmap = sns.heatmap(data = test,
           xticklabels = test.columns, 
           yticklabels = test.index,
           vmin=0,
           cmap="seismic",
           ax=ax,
           linewidth=.5,
            center = 1)

    heatmap.set_xlabel("gRNA", fontsize=10)
    heatmap.set_ylabel("Fraction", fontsize=10)
    heatmap.xaxis.labelpad= 10
    heatmap.yaxis.labelpad= 10
    plt.show()
```

# Aho-Corasick Algorithm, 0 mismatch, no assertion check


```python
ratio_benchmark_heatmap(file="Miniscreen05/Sequence/read_counts(fast_no_assertion).csv")
```

# Aho-Corasick Algorithm, 0 mismatch, assertion check


```python
ratio_benchmark_heatmap(file="Miniscreen05/Sequence/read_counts(fast_yes_assertion).csv")
```

# ARenh

```python
def read_count_heatmap(file):
    df = pd.read_csv(file, index_col=0)
    filter_df = df[np.mod(np.arange(1, df.index.size + 1),9)!=0].iloc[:80, 1:]
    filter_df.rename(columns={'Vec':'Vector'}, inplace=True)
    
    test = filter_df.replace(0,1)

    sns.set(font_scale=2)
    fig, ax = plt.subplots(figsize=(40,15))
    heatmap = sns.heatmap(data = np.log(test),
           xticklabels = "", 
           yticklabels = test.index,
           vmin=0,
           cmap="Reds",
           ax=ax,
           linewidth=.5)

    heatmap.set_xlabel("gRNA", fontsize=20)
    heatmap.set_ylabel("Fraction", fontsize=20)
    heatmap.xaxis.labelpad= 20
    heatmap.yaxis.labelpad= 20
    plt.show()
```

```python
read_count_heatmap("ARenh/read_counts.csv")
```
